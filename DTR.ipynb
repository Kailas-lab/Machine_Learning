```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
X = np.array([[1], [2], [3], [4], [5]])  # Feature
Y = np.array([3, 4, 5, 6, 7])            # Target

# Create the decision tree regressor model
model = DecisionTreeRegressor(random_state=0)

# Fit the model
model.fit(X, Y)

# Take input from user for prediction
user_input = input("Enter comma-separated values for prediction (e.g., 1.5, 2.5, 3.5): ")
X_test = np.array([[float(value)] for value in user_input.split(",")])

# Predict new values
Y_pred = model.predict(X_test)

# Output predictions
print("Predicted values:")
for x, y in zip(X_test.flatten(), Y_pred):
    print(f"Input: {x}, Prediction: {y}")

# Plotting the tree's predictions
plt.scatter(X, Y, color='blue', label="Original Data")  # Original data
plt.scatter(X_test, Y_pred, color='green', label="User Predictions")  # Predicted points
plt.plot(X_test, Y_pred, color='red', label="Prediction Line")  # Prediction line
plt.legend()
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Decision Tree Regression")
plt.show()
```

```python
# Plotting the decision tree structure
from sklearn.tree import plot_tree


plt.figure(figsize=(12, 6))
plot_tree(model, feature_names=["Feature"], filled=True)
plt.title("Decision Tree Structure")
plt.show()
```

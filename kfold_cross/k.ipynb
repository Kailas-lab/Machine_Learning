```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

import numpy as np
from sklearn.datasets import load_digits

digits=load_digits()
```

```python
dir(digits)
```

```python
import pandas as pd


df = pd.DataFrame(digits.data)
df
```

```python
df['target']=digits.target
df
```

```python
X= df.drop(df[['target']],axis='columns')
y=df.target
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)

lr.score(X_test,y_test)
```

```python
svc=SVC()
svc.fit(X_train,y_train)

svc.score(X_test,y_test)
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,y_train)

rf.score(X_test,y_test)
```

```python
from sklearn.model_selection import KFold

kf=KFold(n_splits=3)
kf
```

```python
for train_idex,test_index in kf.split([1,2,3,4,5,6,7,8,9]):
    print(train_idex,test_index)
```

```python
def get_score(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)  
```

```python
#  example

print(get_score(lr,X_train,X_test,y_train,y_test))
print(get_score(rf,X_train,X_test,y_train,y_test))
print(get_score(svc,X_train,X_test,y_train,y_test))
```

![img](77.png)

```python
from sklearn.model_selection import StratifiedKFold

folds= StratifiedKFold(n_splits=3)
folds
```

```python
# collect the scores of the model, because they give diff scores on each iterations(total fold = 3)

score_lr=[]
score_rf=[]
score_svc=[]
```

```python
for train_index, test_index in folds.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    score_lr.append(get_score(lr, X_train, X_test, y_train, y_test))
    score_rf.append(get_score(rf, X_train, X_test, y_train, y_test))
    score_svc.append(get_score(svc, X_train, X_test, y_train, y_test))
```

```python
score_lr
```

```python
score_rf
```

```python
score_svc
```

**we define one function to perform all model and get the scores then checks which is best classifier model for our task**

### instead of doing manaully , we use the cross val score method to get this.

```python
from sklearn.model_selection import cross_val_score

cross_val_score(lr,X,y,cv=3)
```

```python
cross_val_score(rf,X,y,cv=3)
```

```python
cross_val_score(svc,X,y,cv=3)
```

**if we want use parameter tuning and take the average of scores which is best fit**

```python
scores1 = cross_val_score(RandomForestClassifier(n_estimators=5),X, y, cv=10)
np.average(scores1)
```

```python
scores2 = cross_val_score(SVC(gamma='auto'), X, y,cv=3)
np.average(scores2)
```

```python
scores3 = cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'),X, y ,cv=3)
np.average(scores3)
```

## Conclusion

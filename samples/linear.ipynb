```python
from sklearn import linear_model
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv("a2.csv")
a.head()
```

```python
plt.scatter(a.area, a.price, marker="+")
plt.xlabel("area")
plt.ylabel("price")
```

```python
reg = linear_model.LinearRegression()
reg.fit(a[['area']],a['price'])
```

```python
reg.predict([[3300]])
```

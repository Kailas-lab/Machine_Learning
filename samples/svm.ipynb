```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.data
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

add the target column to this df

```python
df['target']=iris.target
df
```

```python
iris.target_names
```

```python
df[df['target']==1].head()
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
df
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
df1.head()
```

```python
import matplotlib.pyplot as plt

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='g',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='r',marker='*')
plt.show()
```

```python

plt.xlabel('patel length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='g',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='r',marker='*')
plt.show()
```

```python
X=df.drop(['target','flower_name'],axis='columns')
X
```

```python
y = df.target
y
```

**Train Using Support Vector Machine (SVM)**

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC

model = SVC()

```

```python
model.fit(X_train,y_train)
```

```python
model.predict([[7.8,9.0,9.5,0.3]])
```

**Tune parameters**

**1. Regularization (C)**

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

**2. Gamma**

```python
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

**3. Kernel**

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```

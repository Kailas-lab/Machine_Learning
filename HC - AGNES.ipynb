```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Sample data
data = np.array([
    [2, 5],   # Point A
    [6, 5],   # Point B
    [2, 4],   # Point C
    [2, 2],   # Point D
    [5, 4],   # Point E
])

# Labels for data points
labels = ['A', 'B', 'C', 'D', 'E']

# Perform Agglomerative Clustering using Scikit-learn
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0, linkage='single')

# Fit the model
model.fit(data)

# Generate the linkage matrix for dendrogram
linkage_matrix = linkage(data, method='single')

# Plot the dendrogram
plt.figure(figsize=(8, 5))
dendrogram(linkage_matrix, labels=labels, color_threshold=3)
plt.title("AGNES Hierarchical Clustering (Dendrogram)")
plt.xlabel("Points")
plt.ylabel("Distance")
plt.show()


```
